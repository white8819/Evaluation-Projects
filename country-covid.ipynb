```python
#to reduce overfitting we drop irrelevant columns
df.drop(['month'],axis=1,inplace=True)
df.drop(['day'],axis=1,inplace=True)
df.drop(['0-14 aged'],axis=1,inplace=True)
df.drop(['15-64 aged'],axis=1,inplace=True)
df.drop(['65+ aged'],axis=1,inplace=True)
```

```python
#X Y Split
x=df.drop(['Death_rate'],axis=1)
y=df['Death_rate']
y=y.values.reshape(-1,1)
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
se=StandardScaler()
dx=se.fit_transform(x)
x=pd.DataFrame(dx,columns=x.columns)
x
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.externals import joblib
```

```python
#Calculating best r score
def max_score(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        r2=r2_score(ytest,ypred)
        if r2>max_score:
            max_score=r2
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
    
```

```python
lr=LinearRegression()
max_score(lr)
```

```python
alpha={'alpha':[.0001,.001,.01,.1,1,10]}
grid=GridSearchCV(estimator=Ridge(),param_grid=alpha)
grid.fit(x,y)
print(grid.best_params_)
```

```python
rd=Ridge(alpha=10)
max_score(rd)
```

```python
alpha={'alpha':[.0001,.001,.01,.1,1,10]}
grid1=GridSearchCV(estimator=Lasso(),param_grid=alpha)
grid1.fit(x,y)
print(grid1.best_params_)
```

```python
ls=Lasso(alpha=0.1)
max_score(ls)
```

```python
alpha={'alpha':[.0001,.001,.01,.1,1,10]}
grid=GridSearchCV(estimator=ElasticNet(),param_grid=alpha)
grid.fit(x,y)
print(grid.best_params_)
```

```python
en=ElasticNet(alpha=0.1)
max_score(en)
```

```python
from sklearn.svm import SVR
para={'kernel':['rbf','linear','poly']}
grid=GridSearchCV(estimator=SVR(),param_grid=para)
grid.fit(x,y)
print(grid.best_params_)
```

```python
svr=SVR(kernel='rbf')
max_score(svr)
```

```python
from sklearn.neighbors import KNeighborsRegressor
#Knr
n_neighbors={'n_neighbors':[2,3,4,5,6,7,8,9]}
grid=GridSearchCV(estimator=KNeighborsRegressor(),param_grid=n_neighbors)
grid.fit(x,y)
print(grid.best_params_)
```

```python
knr=KNeighborsRegressor(n_neighbors=9)
max_score(knr)
```

```python
from sklearn.tree import DecisionTreeRegressor
para={'criterion':['mse','mae','friedman_mse']}
grid=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=para)
grid.fit(x,y)
print(grid.best_params_)
```

```python
dtr=DecisionTreeRegressor(criterion='mse')
max_score(dtr)
```

```python
xgbr=XGBRegressor()
max_score(xgbr)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
para={'n_estimators':[100,200,300,500],'learning_rate':[1,0.1,0.01,0.001]}
grid=GridSearchCV(estimator=GradientBoostingRegressor(),param_grid=para)
grid.fit(x,y)
print(grid.best_params_)
```

```python
gdr=GradientBoostingRegressor(learning_rate=0.1,n_estimators=500)
max_score(gdr)
```

```python
from sklearn.ensemble import RandomForestRegressor
para={'n_estimators':[100,200,300,500]}
grid=GridSearchCV(estimator=RandomForestRegressor(),param_grid=para)
grid.fit(x,y)
print(grid.best_params_)
```

```python
rfr=RandomForestRegressor(n_estimators=500)
max_score(rfr)
```

```python
models=[]
models.append(('Linear Regression',lr,96))
models.append(('Support Vector Regressor',svr,53))
models.append(('Lasso Regression',ls,96))
models.append(('Ridge regression',rd,96))
models.append(('Elastic Net',en,84))
models.append(('KNeighborsRegressor',knr,84))
models.append(('DecisionTreeRegressor',dtr,57))
models.append(('XGBoostRegressor',xgbr,65))
models.append(('GradientBoostingRegressor',gdr,46))
models.append(('RandomForestRegressor',rfr,53))

Model=[]
R2Score=[]
Cvs=[]
MAE=[]
MSE=[]
RMSE=[]
```

```python
for name,model,randomstate in models:
    print('***********',name,'************')
    print('\n')
    Model.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    R2=r2_score(ytest,pred)
    print('R2 score=',R2)
    R2Score.append(R2*100)
    print('\n')
    cvs=cross_val_score(model,x,y,cv=10,scoring='r2').mean()
    print('Cross Validation Score=',cvs)
    Cvs.append(cvs)
    print('\n')
    mae=mean_absolute_error(ytest,pred)
    print('Mean Absolute Error=',mae)
    MAE.append(mae)
    print('\n')
    mse=mean_squared_error(ytest,pred)
    print("Mean Squared Error=",mse)
    MSE.append(mse)
    print('\n')
    rmse=np.sqrt(mean_squared_error(ytest,pred))
    print('Root Mean Squared Error=',rmse)
    RMSE.append(rmse)
```

```python
Result=pd.DataFrame({'Model':Model,'R2_Score':R2Score,'Cross Validation Score':Cvs,'Mean Absolute Error':MAE,'Mean Squared Error':MSE,'Root Mean Squared Error':RMSE})
Result
```

GDR is the best model in terms of crossvalidation score
