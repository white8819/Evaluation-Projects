```python
#x y division of dataset
x=df.drop(['num'],axis=1)
y=df['num']
y=y.values.reshape(-1,1)
y.shape
```

```python
#Scaling data
import sklearn
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
dx=scale.fit_transform(x)
x=pd.DataFrame(dx,columns=x.columns)
x
```

Modelling-Since y variable is categorical classification is used

```python
#Importing Libraries
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
from sklearn.model_selection import cross_val_score
```

```python
# getting the best random state for train test split
def max_score(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        accuracy=accuracy_score(ytest,ypred)
        print('Accuracy score is',accuracy,'with r state',r_state)
        if accuracy>max_score:
            max_score=accuracy
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
```

```python
logr=LogisticRegression()
max_score(logr)
```

```python
#finding best parameters for knn
algo=KNeighborsClassifier()
para={'n_neighbors':[3,4,5,6,7,8,9,10]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
knc=KNeighborsClassifier(n_neighbors=6)
max_score(knc)
```

```python
algo=SVC()
para={'kernel':['rbf','poly','linear'],'C':[1,10]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
svc=SVC(kernel='poly',C=1)
max_score(svc)
```

```python
algo=DecisionTreeClassifier()
para={'criterion':['gini','entropy']}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
max_score(dtc)
```

```python
algo=RandomForestClassifier()
para={'n_estimators':[50,100,150,200,500]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
rfc=RandomForestClassifier(n_estimators=100)
max_score(rfc)
```

```python
ad=AdaBoostClassifier(n_estimators=200,base_estimator=dtc)
max_score(ad)
```

```python
gdb=GradientBoostingClassifier(learning_rate=0.1)
max_score(gdb)
```

```python
models2=[]
models2.append(('Logistic Regression',logr,95))
models2.append(('Support Vector Classifier',svc,77))
models2.append(('KNeighborsClassifier',knc,46))
models2.append(('Decision tree Classifier',dtc,67))
models2.append(('RandomForest Classifier',rfc,71))
models2.append(('AdaBoost Classifier',ad,45))
models2.append(('Gradient Boosting Classifier',gdb,74))

Models=[]
Accuracy=[]
Cvs=[]
```

```python
for name,model,randomstate in models2:
    print('***********',name,'************')
    print('\n')
    Models.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    Ac=accuracy_score(ytest,pred)
    print('Accuracy score=',Ac)
    Accuracy.append(Ac*100)
    print('\n')
    cvs=cross_val_score(model,x,y,cv=5,scoring='accuracy').mean()
    print('Cross Validation Score=',cvs)
    Cvs.append(cvs)
    print('\n')
    Cr=classification_report(ytest,pred)
    print(Cr)
    print('\n')
    cm=confusion_matrix(ytest,pred)
    print(cm)
  
```

```python
Result=pd.DataFrame({'Models':Models,'Accuracy':Accuracy,'Cross Validation Score':Cvs})
Result
```
