```python
df.skew()
dfn=df.drop(['y'],axis=1)
```

```python
#skewness treatment
for i in dfn.columns:
    if df.skew().loc[i]>0.55:
        df[i]=df[i].apply(lambda x:np.log(x+1))
```

```python
#X y division of data
x=df.drop(['y'],axis=1)
y=df['y']
y=y.values.reshape(-1,1)
y.shape
```

```python
#scaling the data
import sklearn
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
dx=scale.fit_transform(x)
x=pd.DataFrame(dx,columns=x.columns)
x
```

Modelling the data- Since y variable is categorical classification and logistic regression is used.

```python
#Importing Libraries
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
from sklearn.model_selection import cross_val_score
```

```python
# getting the best random state for train test split
def max_score(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        accuracy=accuracy_score(ytest,ypred)
        if accuracy>max_score:
            max_score=accuracy
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
```

```python
logr=LogisticRegression()
max_score(logr)
```

```python
#finding best parameters for knn
algo=KNeighborsClassifier()
para={'n_neighbors':[3,5,7,9,11]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
knc=KNeighborsClassifier(n_neighbors=11)
max_score(knc)
```

```python
algo=SVC()
para={'kernel':['rbf','poly','linear'],'C':[1,10]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
svc=SVC(C= 10, kernel= 'rbf')
max_score(svc)
```

```python
algo=DecisionTreeClassifier()
para={'criterion':['gini','entropy']}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
dtc=DecisionTreeClassifier(criterion='entropy')
max_score(dtc)
```

```python
algo=RandomForestClassifier()
para={'n_estimators':[50,100,150,200,500]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x,y)

print(grid.best_params_)
```

```python
rfc=RandomForestClassifier(n_estimators=150)
max_score(rfc)
```

```python
models=[]
models.append(('Logistic Regression',logr,93))
models.append(('Support Vector Classifier',svc,93))
models.append(('KNeighborsClassifier',knc,85))
models.append(('Decision tree Classifier',dtc,75))
models.append(('RandomForest Classifier',rfc,93))
#models.append(('AdaBoost Classifier',ad,93))
models.append(('Gradient Boosting Classifier',gdb,93))

Models=[]
Accuracy=[]
Cvs=[]
Roc_score=[]
```

```python
for name,model,randomstate in models:
    print('***********',name,'************')
    print('\n')
    Models.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    Ac=accuracy_score(ytest,pred)
    print('Accuracy score=',Ac)
    Accuracy.append(Ac*100)
    print('\n')
    cvs=cross_val_score(model,x,y,cv=5,scoring='accuracy').mean()
    print('Cross Validation Score=',cvs)
    Cvs.append(cvs)
    print('\n')
    Cr=classification_report(ytest,pred)
    print(Cr)
    print('\n')
    cm=confusion_matrix(ytest,pred)
    print(cm)
    fpr,tpr,threshold=roc_curve(ytest,pred)
    roc_auc=auc(fpr,tpr)
    print('Roc score=',roc_auc)
    Roc_score.append(roc_auc)
    print('\n')
    
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(fpr,tpr,label='AUC=%0.2f'%roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True Positive Rate')
    
    print('\n\n')
```

```python
Result=pd.DataFrame({'Models':Models,'Accuracy':Accuracy,'Cross Validation Score':Cvs,'Roc_Score':Roc_score})
Result
```
