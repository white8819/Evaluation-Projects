```python
#Categorising the x and y variables
x=df.iloc[:,1:18]
y=df['Wins']
```

```python
y=y.values.reshape(-1,1)
y.shape
```

```python
#Scaling the data
import sklearn
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
dx=scale.fit_transform(x)
x=pd.DataFrame(dx,columns=x.columns)
x
```

```python
#PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
pca_x=pca.fit_transform(x)
```

Now since the y variable is continuous, Linear Regression is the preferred model to use.

Modelling

```python
#Importing libraries
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import cross_val_score
```

```python
# getting the best random state for train test split
def max_score(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        r2=r2_score(ytest,ypred)
        print('r2 score is',r2,'with r state',r_state)
        if r2>max_score:
            max_score=r2
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
   
```

```python
# Using GridSearchCV to find the best parameters
#Ridge Reg
from sklearn.model_selection import GridSearchCV
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid=GridSearchCV(estimator=Ridge(),param_grid=alpha)
grid.fit(x,y)
print(grid.best_params_)
```

```python
rd=Ridge(alpha=1)
max_score(rd)
```

```python
#Lasso Reg
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid1=GridSearchCV(estimator=Lasso(),param_grid=alpha)
grid1.fit(x,y)
print(grid1.best_params_)
```

```python
ls=Lasso(alpha=1)
max_score(ls)
```

```python
#ElasticNet
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid=GridSearchCV(estimator=ElasticNet(),param_grid=alpha)
grid.fit(x,y)
print(grid.best_params_)
```

```python
en=ElasticNet(alpha=0.01)
max_score(en)
```

```python
#SVR
from sklearn.svm import SVR
svr=SVR(kernel='linear')
max_score(svr)
```

```python
max_score(LinearRegression())
```

```python
#KNR
n_neighbors={'n_neighbors':[2,3,4,5,6,7,8,9]}
grid=GridSearchCV(estimator=KNeighborsRegressor(),param_grid=n_neighbors)
grid.fit(x,y)
print(grid.best_params_)
```

```python
knr=KNeighborsRegressor(n_neighbors=4)
max_score(knr)
```

```python
lr=LinearRegression()
models=[]
models.append(('Linear Regression',lr,86))
models.append(('Lasso Regression',ls,82))
models.append(('Ridge regression',rd,69))
models.append(('Elastic Net',en,69))
models.append(('Support Vector Regressor',svr,75))
models.append(('KNeighborsRegressor',knr,82))

Model=[]
R2Score=[]
Cvs=[]
MAE=[]
MSE=[]
RMSE=[]
```

```python
for name,model,randomstate in models:
    print('***********',name,'************')
    print('\n')
    Model.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    R2=r2_score(ytest,pred)
    print('R2 score=',R2)
    R2Score.append(R2*100)
    print('\n')
    cvs=cross_val_score(model,x,y,cv=5,scoring='r2').mean()
    print('Cross Validation Score=',cvs)
    Cvs.append(cvs)
    print('\n')
    mae=mean_absolute_error(ytest,pred)
    print('Mean Absolute Error=',mae)
    MAE.append(mae)
    print('\n')
    mse=mean_squared_error(ytest,pred)
    print("Mean Squared Error=",mse)
    MSE.append(mse)
    print('\n')
    rmse=np.sqrt(mean_squared_error(ytest,pred))
    print('Root Mean Squared Error=',rmse)
    RMSE.append(rmse)
    
    
```

```python
Model
```

```python
Result=pd.DataFrame({'Model':Model,'R2_Score':R2Score,'Cross Validation Score':Cvs,'Mean Absolute Error':MAE,'Mean Squared Error':MSE,'Root Mean Squared Error':RMSE})
Result
```

Linear Regression model is the best model based on R2 score and error values.

```python
#modelling on PCA data

models2=[]
models2.append(('Linear Regression',lr,69))
models2.append(('Lasso Regression',ls2,69))
models2.append(('Ridge regression',rd,69))
models2.append(('Elastic Net',en2,69))
models2.append(('Support Vector Regressor',svr,75))
models2.append(('KNeighborsRegressor',knr,82))

Model_pca=[]
R2Score_pca=[]
Cvs_pca=[]
MAE_pca=[]
MSE_pca=[]
RMSE_pca=[]
```
