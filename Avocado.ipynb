Modelling-First Average price is taken as the y variable, which is continuous so Linear Regression is preferred.

```python
df
```

```python
#x and y division of data
x1=df.iloc[:,1:11]
y1=df['AveragePrice']
y1=y1.values.reshape(-1,1)
y1.shape
```

```python
#Scaling the data
import sklearn
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
dx=scale.fit_transform(x1)
x1=pd.DataFrame(dx,columns=x1.columns)
x1
```

```python
#Importing libraries
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import cross_val_score
```

```python
# getting the best random state for train test split
def max_score(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x1,y1,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        r2=r2_score(ytest,ypred)
        print('r2 score is',r2,'with r state',r_state)
        if r2>max_score:
            max_score=r2
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
   
```

```python
max_score(LinearRegression())
```

```python
# Using GridSearchCV to find the best parameters
#Ridge Reg
from sklearn.model_selection import GridSearchCV
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid=GridSearchCV(estimator=Ridge(),param_grid=alpha)
grid.fit(x1,y1)
print(grid.best_params_)
```

```python
rd=Ridge(alpha=1)
max_score(rd)
```

```python
#Lasso Reg
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid1=GridSearchCV(estimator=Lasso(),param_grid=alpha)
grid1.fit(x1,y1)
print(grid1.best_params_)
```

```python
ls=Lasso(alpha=0.01)
max_score(ls)
```

```python
#ElasticNet
alpha={'alpha':[1,0.1,0.01,0.001,0.0001,0]}
grid=GridSearchCV(estimator=ElasticNet(),param_grid=alpha)
grid.fit(x1,y1)
print(grid.best_params_)
```

```python
en=ElasticNet(alpha=0.1)
max_score(en)
```

```python
#SVR
from sklearn.svm import SVR
svr=SVR(kernel='linear')
max_score(svr)
```

```python
#KNR
n_neighbors={'n_neighbors':[2,3,4,5,6,7,8,9]}
grid=GridSearchCV(estimator=KNeighborsRegressor(),param_grid=n_neighbors)
grid.fit(x1,y1)
print(grid.best_params_)
```

```python
knr=KNeighborsRegressor(n_neighbors=9)
max_score(knr)
```

```python
#DTR
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
max_score(dtr)
```

```python
#Random forest regressor
from sklearn.ensemble import RandomForestRegressor
para={'n_estimators':[10,50,100,150,200,500],'criterion':['mse','mae']}
grid=GridSearchCV(estimator=RandomForestRegressor(),param_grid=para)
grid.fit(x1,y1)
print(grid.best_params_)
```

```python
rfr=RandomForestRegressor(criterion='mse',n_estimators=150)
max_score(rfr)
```

```python
lr=LinearRegression()
models=[]
models.append(('Linear Regression',lr,99))
models.append(('Lasso Regression',ls,46))
models.append(('Ridge regression',rd,99))
models.append(('Elastic Net',en,44))
models.append(('Support Vector Regressor',svr,99))
models.append(('KNeighborsRegressor',knr,60))
models.append(('Decision tree Regressor',dtr,43))
models.append(('RandomForest Regressor',rfr,43))

Model=[]
R2Score=[]
Cvs=[]
MAE=[]
MSE=[]
RMSE=[]
```

```python
for name,model,randomstate in models:
    print('***********',name,'************')
    print('\n')
    Model.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x1,y1,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    R2=r2_score(ytest,pred)
    print('R2 score=',R2)
    R2Score.append(R2*100)
    print('\n')
    cvs=cross_val_score(model,x1,y1,cv=5,scoring='r2').mean()
    print('Cross Validation Score=',cvs)
    Cvs.append(cvs)
    print('\n')
    mae=mean_absolute_error(ytest,pred)
    print('Mean Absolute Error=',mae)
    MAE.append(mae)
    print('\n')
    mse=mean_squared_error(ytest,pred)
    print("Mean Squared Error=",mse)
    MSE.append(mse)
    print('\n')
    rmse=np.sqrt(mean_squared_error(ytest,pred))
    print('Root Mean Squared Error=',rmse)
    RMSE.append(rmse)
    
```

```python
Result=pd.DataFrame({'Model':Model,'R2_Score':R2Score,'Cross Validation Score':Cvs,'Mean Absolute Error':MAE,'Mean Squared Error':MSE,'Root Mean Squared Error':RMSE})
Result
```

RandomForest Regressor is the best model in terms of R2 score an error values.

```python
#x and y division of data
x2=df.drop(['region'],axis=1)
y2=df['region']
y2=y2.values.reshape(-1,1)
y2.shape
```

```python
#Scaling the data
scale=StandardScaler()
dx1=scale.fit_transform(x2)
x2=pd.DataFrame(dx1,columns=x2.columns)
x2
```

```python
#Importing Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
```

```python
# getting the best random state for train test split
def max_score2(model):
    max_score=0
    for r_state in range(42,100):
        xtrain,xtest,ytrain,ytest=train_test_split(x2,y2,random_state=r_state,test_size=0.20)
        model.fit(xtrain,ytrain)
        ypred=model.predict(xtest)
        accuracy=accuracy_score(ytest,ypred)
        print('Accuracy score is',accuracy,'with r state',r_state)
        if accuracy>max_score:
            max_score=accuracy
            final_r_state=r_state
    print(max_score,'is max accuracy against r_state',final_r_state)
```

```python
logr=LogisticRegression()
max_score2(logr)
```

```python
dtc=DecisionTreeClassifier(criterion='entropy')
max_score2(dtc)
```

```python
algo=RandomForestClassifier()
para={'n_estimators':[50,100,150,200,500]}
grid=GridSearchCV(estimator=algo,param_grid=para)
grid.fit(x2,y2)

print(grid.best_params_)
```

```python
rfc=RandomForestClassifier(n_estimators=200)
max_score2(rfc)
```

```python
models2=[]
models2.append(('Logistic Regression',logr,51))
models2.append(('Support Vector Classifier',svc,58))
models2.append(('KNeighborsClassifier',knc,52))
models2.append(('Decision tree Classifier',dtc,64))
models2.append(('RandomForest Classifier',rfc,58))

Models=[]
Accuracy=[]
Cvs2=[]
```

```python
for name,model,randomstate in models2:
    print('***********',name,'************')
    print('\n')
    Models.append(name)
    xtrain,xtest,ytrain,ytest=train_test_split(x2,y2,random_state=randomstate,test_size=0.20)
    model.fit(xtrain,ytrain)
    print(model)
    pred=model.predict(xtest)
    print('\n')
    Ac=accuracy_score(ytest,pred)
    print('Accuracy score=',Ac)
    Accuracy.append(Ac*100)
    print('\n')
    cvs=cross_val_score(model,x2,y2,cv=5,scoring='accuracy').mean()
    print('Cross Validation Score=',cvs)
    Cvs2.append(cvs)
    print('\n')
    Cr=classification_report(ytest,pred)
    print(Cr)
    print('\n')
    cm=confusion_matrix(ytest,pred)
    print(cm)
```

```python
Result2=pd.DataFrame({'Models':Models,'Accuracy':Accuracy,'Cross Validation Score':Cvs2})
Result2
```

The support vector classifier is the best model based on accuracy and cross validation score.
